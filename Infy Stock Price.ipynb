```python
import graphlab
```

```python
url = '/Users/Aniket/anaconda/infy.csv'
infy = graphlab.SFrame.read_csv(url)
infy['Date'] = infy['Date'].str_to_datetime('%d/%m/%y')
```

```python
infy.head()
```

```python
graphlab.set_runtime_config('GRAPHLAB_DEFAULT_NUM_PYLAMBDA_WORKERS', 4)
```

```python
graphlab.canvas.set_target('ipynb')
infy.show(view="Line Chart", x="Date", y="Close")
```

```python
infy.show()
```

```python
graphlab.canvas.set_target('ipynb')
infy.show(view="Scatter Plot", x="Revenue", y="Close")
train_data,test_data = infy.random_split(0.8,seed=0)
```

```python
train_data
```

```python
infy_model = graphlab.linear_regression.create(train_data, target='Close', features=['Revenue'])
```

```python
print test_data['Close'].mean()
```

```python
print infy_model.evaluate(test_data)
```

```python
infy_model.get('coefficients')
```
