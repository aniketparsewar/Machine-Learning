```python
def get_residual_sum_of_squares(model, data, outcome):
    # First get the predictions
    predictions = model.predict(data)
    
    # Then compute the residuals/errors
    residual = outcome - predictions
    
    # Then square and add them up
    residual_squared = residual * residual
    RSS = residual_squared.sum()

    return(RSS)   
```

```python

# Compute the RSS on TRAINING data for each of the three models and record the values:
print get_residual_sum_of_squares(model_one, train_data, train_data['price'])
print get_residual_sum_of_squares(model_two, train_data, train_data['price'])
print get_residual_sum_of_squares(model_three, train_data, train_data['price'])
```

```python
# Compute the RSS on TESTING data for each of the three models and record the values:
print get_residual_sum_of_squares(model_one, test_data, test_data['price'])
print get_residual_sum_of_squares(model_two, test_data, test_data['price'])
print get_residual_sum_of_squares(model_three, test_data, test_data['price'])
```
